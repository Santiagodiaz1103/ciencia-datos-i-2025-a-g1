# **Introduccion**
El presente manual tiene como objetivo explicar y demostrar, mediante ejemplos pr치cticos en Python, el uso de funciones de tiempo y jerarqu칤a de fechas. El manejo adecuado de datos temporales es crucial en ciencia de datos.



```python
import pandas as pd
# Crear un rango de fechas
fechas = pd.date_range(start='2023-01-01', periods=10, freq='D')
print(fechas)
# Extraer elementos
df = pd.DataFrame({'fechas': fechas})
df['a침o'] = df['fechas'].dt.year
df['mes'] = df['fechas'].dt.month
df['d칤a'] = df['fechas'].dt.day
df['d칤a_semana'] = df['fechas'].dt.day_name()
df
# Agregar m치s niveles jer치rquicos
df['semana'] = df['fechas'].dt.isocalendar().week
df['trimestre'] = df['fechas'].dt.quarter

df
```

### **Analisis**
La jerarqu칤a de fechas se refiere a la descomposici칩n de una fecha en sus diferentes niveles de granularidad: a침o, trimestre, mes, semana, d칤a, hora, minuto, etc. Esta jerarqu칤a permite organizar y analizar los datos de forma m치s eficiente, seg칰n el nivel de detalle requerido.

```python
import numpy as np
import pandas as pd

# Crear un DataFrame con fechas diarias y ventas aleatorias
df = pd.DataFrame({
    'fecha': pd.date_range(start='2023-01-01', periods=90, freq='D'),  # Genera 90 fechas diarias desde el 1 de enero de 2023
    'ventas': np.random.randint(100, 1000, size=90)  # Genera 90 valores aleatorios entre 100 y 999 para la columna 'ventas'
})

# Crear una nueva columna llamada 'mes' a partir de la columna 'fecha'
# Se utiliza .dt.to_period('M') para convertir la fecha al periodo mensual (por ejemplo, 2023-01)
df['mes'] = df['fecha'].dt.to_period('M')

# Agrupar los datos por la columna 'mes' y sumar las ventas de cada mes
ventas_mensuales = df.groupby('mes')['ventas'].sum()

# Mostrar el resultado de ventas mensuales
ventas_mensuales
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Crear un DataFrame simulado con fechas diarias y ventas aleatorias
df = pd.DataFrame({
    'fecha': pd.date_range(start='2023-01-01', periods=180, freq='D'),  # 180 d칤as (~6 meses)
    'ventas': np.random.randint(100, 1000, size=180)
})

# 2. Agregar columna de trimestre
df['trimestre'] = df['fecha'].dt.to_period('Q')  # Ej: '2023Q1', '2023Q2'

# 3. Agrupar por trimestre y calcular estad칤sticas descriptivas
estadisticas_trimestrales = df.groupby('trimestre')['ventas'].describe()
print("游늵 Estad칤sticas por trimestre:")
print(estadisticas_trimestrales)

# 4. Agregar columna de mes
df['mes'] = df['fecha'].dt.to_period('M')  # Ej: '2023-01', '2023-02'

# 5. Calcular ventas promedio por mes
ventas_promedio_mensuales = df.groupby('mes')['ventas'].mean()

# 6. Visualizar las ventas promedio por mes
plt.figure(figsize=(10, 5))
ventas_promedio_mensuales.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Ventas Promedio por Mes')
plt.xlabel('Mes')
plt.ylabel('Ventas Promedio')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
